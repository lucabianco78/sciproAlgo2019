## Backtracking code

Finding all subsets of integers in ${0,..., n-1}$:

```python
%reset -s -f 

def process_solution(S):
    for i in range(len(S)):
        print(S[i], end = " ")
    print("")
    return False

def subsets(S,n,i):
    print("subsets({},{},{})".format(S,n,i))
    C = [1, 0] if i<n else []
    #print(C)
    for c in C:
        S[i] = c
        #print("S now: {}, c:{}".format(S,c))
        if i == n-1:
            print("S:{} c:{} i:{}".format(S,c,i))
            if process_solution(S):
                return True
        else:
            #print("Calling:  subsets({},{},{})".format(S,n,i+1))
            subsets(S,n,i+1)
    return False

n = 5
S = [0]*n
subsets(S,n,0)
```

```python
%reset -s -f 

def process_solution(S):
    for i in range(len(S)):
        print(S[i], end = " ")
    print("")
    return False

def subsets(S,n,i, counter):
    
    print("subsets({},{},{},{})".format(S,n,i, counter))
    C = [1, 0] if i<n else []
    #print(C)
    for c in C:
        S[i] = c
        #print("S now: {}, c:{}".format(S,c))

        if i == n-1:
            print("S:{} c:{} i:{} counter:{}".format(S,c,i,counter))
            if process_solution(S):
                return True
        else:
            #print("Calling:  subsets({},{},{})".format(S,n,i+1))
            subsets(S,n,i+1, counter)
        counter += 1
    return False

n = 5
S = [0]*n
subsets(S,n,0,0)
```
